```python
import math
def akar_persamaan_kuadrat(a, b, c):
  diskriminan = b**2 - 4*a*c
  if diskriminan < 0:
    return "Tidak ada akar real"
  else:
    akar1 = (-b + math.sqrt(diskriminan)) / (2*a)
    akar2 = (-b - math.sqrt(diskriminan)) / (2*a)
    return akar1, akar2
a = 1
b = -3
c = -810
panjang = akar_persamaan_kuadrat(a, b, c)
panjang = max(panjang)
lebar = panjang - 3
print("Panjang lahan:", panjang, "meter")
print("Lebar lahan:", lebar, "meter")
```
